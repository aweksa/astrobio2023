```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
```

```python
planeta1 = pd.read_csv(r'planeta1.csv', header = None)
planeta2 = pd.read_csv(r'planeta2.csv', header = None)
planeta3 = pd.read_csv(r'planeta3.csv', header = None)
planeta4 = pd.read_csv(r'planeta4.csv', header = None)
```

```python
plt.scatter(planeta1[0], planeta1[1])
```

```python
def blackbody_lam(lam, T):
    I=1.910e+04 / (lam**5 * (np.exp(14387.7 / (lam*T)) - 1))
    return I
def func(wa, T1):
    return blackbody_lam(wa, T1)

popt, pcov = curve_fit(func, planeta1[0], planeta1[1]/5000, p0=350)

sigmaT1 = np.sqrt(pcov)
# you can print it
print (' T1 = %.2f +/- %.2f' % (popt, sigmaT1))
# calculate the modeled curve, w is
# exoplanet wavelenght
ybest = blackbody_lam(planeta1[0], popt)
# the real spectrum
plt.plot(planeta1[0],planeta1[1],'r', label = 'data')
# modeled spectrum, adjust units
plt.plot(planeta1[0],ybest*5000,'g', label = 'planck')
plt.legend()
```
